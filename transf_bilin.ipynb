```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
z, p = sp.symbols('z, p')
Te, r, omega, Omega = sp.symbols('T_e, r, omega, Omega', real=True)

p = (2/Te)*(1-z**-1)/(1+z**-1)

display(Math('p = '+sp.latex(p)))
```

```python
'p = '+sp.latex(p)
```

```python
pz = p.subs(z, r*sp.exp(sp.I*Omega*Te))
display(Math('p = '+sp.latex(pz)))
```

```python
'p = '+sp.latex(pz)
```

```python
Rep = sp.simplify(pz.as_real_imag()[0])
Imp = sp.simplify(pz.as_real_imag()[1])
display(Math('\Re(p) = '+sp.latex(Rep)))
display(Math('\Im(p) = '+sp.latex(Imp)))
```

```python
'\Re(p) = '+sp.latex(Rep)
```

```python
'\Im(p) = '+sp.latex(Imp)
```

```python
display(Math('\Re(p) = '+sp.latex(Rep.subs(r, 1))))
display(Math('\Im(p) = '+sp.latex(Imp.subs(r, 1))))
```

```python
Imp = sp.simplify(Imp.subs(r, 1))
display(Math(sp.latex(Imp)))
```

```python
Om = sp.solve(Imp - omega, Omega)[0]
display(Math('\Omega = '+sp.latex(Om)))
```

```python
'\Omega = '+sp.latex(Om)
```

```python
om = sp.solve(Om-Omega, omega)[0]
display(Math('\omega = '+sp.latex(om)))
```

```python
'\omega = '+sp.latex(om)
```
